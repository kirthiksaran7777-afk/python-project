```python
import numpy as np
import pandas as pd

data = pd.read_csv("supermarket_sales.csv")

quantity = np.array(data["Quantity"])
unit_price = np.array(data["Unit price"])
total_sales = np.array(data["Total"])
profit = np.array(data["gross income"])

print("Total Revenue:", np.sum(total_sales))
print("Average Sales:", np.mean(total_sales))

print("Mean:", np.mean(total_sales))
print("Median:", np.median(total_sales))
print("Variance:", np.var(total_sales))
print("Standard Deviation:", np.std(total_sales))

revenue = quantity * unit_price
print("Calculated Revenue:", revenue)

print("Total Profit:", np.sum(profit))
print("Max Sale:", np.max(total_sales))
print("Min Sale:", np.min(total_sales))
```
